## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """
SELECT *
FROM DatabaseDesireNextYear
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)

df_05 = pd.DataFrame(df["DatabaseDesireNextYear"].value_counts())
df_05 = df_05.sort_values('DatabaseDesireNextYear', ascending=False).head()
#df_05 = df_05.set_index('DatabaseDesireNextYear')

#plot
df_05["DatabaseDesireNextYear"].plot(kind='pie',autopct='%1.1f%%', labels=None)
plt.title("top 5 databases that respondents wish to learn next year")
plt.legend(labels=df_05.index, loc=(1, 0.6))
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT *
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)

df_sc = df[df.Age<=35]
df_sc =df_sc[df_sc.Age>=30]
df_sc =df_sc.groupby('Age')[["WorkWeekHrs","CodeRevHrs"]].median()

#plot
df_sc.plot(kind='bar', stacked=True)
plt.legend(loc=(1.05, 0.7))
plt.xticks(rotation = 0)
alpha =0.05
plt.grid(True)
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT *
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)

df_c = df[df.Age<=60]
df_c =df_c[df_c.Age>=45]
df_c =df_c.groupby('Age')["ConvertedComp"].median()

#plot
df_c.plot(kind='line')

plt.xlabel("Age")
plt.ylabel("ConvertedComp")
plt.title("ConvertedComp by Age")
plt.legend(["ConvertedComp"], loc=(1.05, 0.9))
alpha =0.05
plt.grid(True)
plt.show()
```

### Bar Chart
